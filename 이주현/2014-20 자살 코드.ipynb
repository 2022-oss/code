```python
import csv
import matplotlib.pyplot as plt

f = open('221126.csv')
data = csv.reader(f)

cuz = next(data)

per = []


for row in data:
    if '총합' in row[0]:
        for i in row[1:]:
            per.append(int(i))


plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(14,7))
plt.bar(cuz[1:], per, color='gray')
plt.title('2014-2022 원인별 자살 인원 총합 (출처 - 경찰청 통계연보)')
plt.xlabel('자살 원인')
plt.ylabel('자살 인원')
plt.show()
```
