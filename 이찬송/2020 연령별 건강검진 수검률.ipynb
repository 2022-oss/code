```python
import matplotlib.pyplot as plt
import csv

f = open('2020 연령별 최종.csv')
data = csv.reader(f)

result = []

header = next(data)

for row in data:
    for i in row[1:16]:
        result.append(int(i.replace(',',"")))
        
#print(result)
        
    
f.close()

plt.figure(figsize=(10,2))
plt.rc('font', family='Malgun Gothic')
plt.title('2020 연령별 건강검진 수검률(출처-국민건강보험)')
plt.plot(['~19','20~24','25~29','30~34','35~39','40~45','45~49','50~54','55~59',
          '60~64','65~67','70~74','75~79','80~84','85~'], result)
plt.xlabel('연령 구분(단위:세)')
plt.ylabel('수검인원(단위:백만)')
plt.grid(True)
plt.show()
```
